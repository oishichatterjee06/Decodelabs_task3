```python
import pandas as pd
df = pd.read_csv("Dataset for Data Analytics week 3 - Sheet1.csv")
df.head()
```

```python
import sqlite3
conn = sqlite3.connect('orders.db')
df.to_sql('orders', conn, if_exists='replace', index=False)
```

```python
query1 = """
SELECT *
FROM orders;
"""

pd.read_sql(query1, conn)
```

```python
query2 = """
SELECT OrderID, Product, Quantity, TotalPrice
FROM orders;
"""

pd.read_sql(query2, conn)
```

```python
query3 = """
SELECT *
FROM orders
WHERE TotalPrice > 2000;
"""

pd.read_sql(query3, conn)
```

```python
query4= """
SELECT *
FROM orders
WHERE OrderStatus = 'Delivered';
"""

pd.read_sql(query4, conn)
```

```python
query5 = """
SELECT *
FROM orders
WHERE PaymentMethod = 'Online';
"""

pd.read_sql(query5, conn)
```

```python
query6 = """
SELECT OrderID, Product, TotalPrice
FROM orders
ORDER BY TotalPrice DESC;
"""

pd.read_sql(query6, conn)
```

```python
query7 = """
SELECT OrderID, Product, TotalPrice
FROM orders
ORDER BY TotalPrice ASC;
"""

pd.read_sql(query7, conn)
```

```python
query8 = """
SELECT COUNT(*) AS Total_Orders
FROM orders;
"""

pd.read_sql(query8, conn)
```

```python
query9 = """
SELECT SUM(TotalPrice) AS Total_Revenue
FROM orders;
"""

pd.read_sql(query9, conn)
```

```python
query10 = """
SELECT AVG(TotalPrice) AS Average_Order_Value
FROM orders;
"""

pd.read_sql(query10, conn)


```

```python
query11 = """
SELECT Product,
       COUNT(*) AS Number_of_Orders
FROM orders
GROUP BY Product
ORDER BY Number_of_Orders DESC;
"""

pd.read_sql(query11, conn)
```

```python
query12 = """
SELECT Product,
       SUM(TotalPrice) AS Revenue
FROM orders
GROUP BY Product
ORDER BY Revenue DESC;
"""

pd.read_sql(query12, conn)
```

```python
query13 = """
SELECT PaymentMethod,
       COUNT(*) AS Orders
FROM orders
GROUP BY PaymentMethod;
"""

pd.read_sql(query13, conn)
```

```python
query14 = """
SELECT PaymentMethod,
       SUM(TotalPrice) AS Revenue
FROM orders
GROUP BY PaymentMethod
ORDER BY Revenue DESC;
"""

pd.read_sql(query14, conn)
```

```python
query15 = """
SELECT OrderStatus,
       COUNT(*) AS Orders
FROM orders
GROUP BY OrderStatus
ORDER BY Orders DESC;
"""

pd.read_sql(query15, conn)
```

```python
query16 = """
SELECT ReferralSource,
       SUM(TotalPrice) AS Revenue
FROM orders
GROUP BY ReferralSource
ORDER BY Revenue DESC;
"""

pd.read_sql(query16, conn)
```

```python
query17 = """
SELECT Product,
       AVG(TotalPrice) AS Avg_Order_Value
FROM orders
GROUP BY Product
ORDER BY Avg_Order_Value DESC;
"""

pd.read_sql(query17, conn)
```

```python
query18 = """
SELECT OrderID,
       Product,
       TotalPrice
FROM orders
ORDER BY TotalPrice DESC
LIMIT 5;
"""

pd.read_sql(query18, conn)
```

```python
query19 = """
SELECT COUNT(*) AS Coupon_Users
FROM orders
WHERE CouponCode IS NOT NULL;
"""

pd.read_sql(query19, conn)
```

```python
query20 = """
SELECT Product,
       SUM(TotalPrice) AS Revenue
FROM orders
GROUP BY Product
HAVING SUM(TotalPrice) > 150000
ORDER BY Revenue DESC;
"""

pd.read_sql(query20, conn)
```
